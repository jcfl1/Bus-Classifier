### The best bus classifier you have ever seen!

Welcome to the bus classifier web application! Here you can play and test with this neural network in order to classify buses



```python
from fastai.vision.all import *
from fastai.vision.widgets import *
```

```python
path = Path()
learn_inf = load_learner(path/'export.pkl', cpu=True)
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
def on_data_change(change):
    lbl_pred.value = ''
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
```

```python
btn_upload.observe(on_data_change, names=['data'])
```

### How to use the Bus Classifier?

It's really simple! You just need to **upload** a bus image, and then it will automatically give you the prediction about it

P.S.: Bus classifier is still in early versions and can only classify a bus in three categories: "london bus", "school bus" or "minibus"

Start trying the application by uploading a image clicking in this button below:

```python
display(VBox([widgets.Label('Select your bus!'), btn_upload, out_pl, lbl_pred]))
```
